# IMPORT LIBRARIES

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Better visualization style
sns.set_style("whitegrid")
```

# LOAD DATASET

```python
df = pd.read_csv("C:/Users/sarth/OneDrive/Desktop/task_02/Amazon.csv")

# display top 5 rows
df.head()
```

```python
# check dataset shape
print("Shape of dataset:", df.shape)
```

```python
# Check column names
print("Columns:", df.columns)
```

```python
# Check data types
df.info()
```

# HANDLE MISSING VALUES

```python
# Check missing values
print(df.isnull().sum())

# Drop missing values
df.dropna(inplace=True)
```

```python
# Verify again
print("After cleaning:")
print(df.isnull().sum())
```

# DATA CLEANING

```python
# Convert date column
df['OrderDate'] = pd.to_datetime(df['OrderDate'], dayfirst=True, errors='coerce')
```

```python
# Standardize column names
df.columns = df.columns.str.lower().str.replace(" ", "_")
```

```python
# Check cleaned data
df.head()
```

# DESCRIPTIVE STATISTICS

```python
# Summary statistics
df.describe()
```

# SALES DISTRIBUTION

```python
# Histogram of sales
plt.figure(figsize=(8,5))
sns.histplot(df['totalamount'], bins=30)

plt.title("Sales Distribution")
plt.xlabel("Sales Amount")
plt.ylabel("Frequency")
plt.show()
```

# MONTHLY SALES ANALYSIS

```python
# Set date as index
df.set_index('orderdate', inplace=True)
```

```python
# Monthly sales
monthly_sales = df['totalamount'].resample('ME').sum()
```

```python
# Plot monthly sales
plt.figure(figsize=(10,5))
monthly_sales.plot()

plt.title("Monthly Sales Trend")
plt.xlabel("Date")
plt.ylabel("Total Sales")
plt.show()
```

# TOP PRODUCTS

```python
# Top 5 products
top_products = df.groupby('productname')['totalamount'].sum().sort_values(ascending=False).head(5)
```

```python
# Plot
plt.figure(figsize=(8,5))
top_products.plot(kind='bar')

plt.title("Top 5 Products")
plt.xlabel("Product")
plt.ylabel("Sales")
plt.show()
```

# CATEGORY ANALYSIS

```python
# Sales by category
category_sales = df.groupby('category')['totalamount'].sum()
```

```python
# Pie chart
plt.figure(figsize=(7,7))
category_sales.plot(kind='pie', autopct='%1.1f%%')

plt.title("Sales by Category")
plt.ylabel("")
plt.show()
```

# MOVING AVERAGE (TREND)

```python
# Calculate moving average
moving_avg = monthly_sales.rolling(window=3).mean()
```

```python
# Plot
plt.figure(figsize=(10,5))
plt.plot(monthly_sales, label='Actual Sales')
plt.plot(moving_avg, label='Moving Average')

plt.title("Sales Trend with Moving Average")
plt.legend()
plt.show()
```

# SIMPLE FORECASTING

```python
# Take last moving average value
last_value = moving_avg.iloc[-1]
```

```python
# Predict next 3 months
forecast = [last_value, last_value, last_value]
```

```python
# Future dates
future_dates = pd.date_range(start=monthly_sales.index[-1], periods=4, freq='ME')[1:]

forecast_series = pd.Series(forecast, index=future_dates)
```

```python
# Plot forecast
plt.figure(figsize=(10,5))

plt.plot(monthly_sales, label='Actual')
plt.plot(forecast_series, label='Forecast', linestyle='--')

plt.title("Simple Sales Forecast")
plt.legend()
plt.show()
```
